```python
from __future__ import print_function
import os
import numpy as np
np.random.seed(1337)
import sys
import codecs

print(sys.getdefaultencoding())

BASE_DIR = ''
GLOVE_DIR = BASE_DIR + '/Users/kimardenmiller/Documents/Local Data/xData/glove.6B/'

print('Indexing word vectors.')

embeddings_index = {}
f = codecs.open(os.path.join(GLOVE_DIR, 'testgood.txt'))
type(f)
# f = open(os.path.join(GLOVE_DIR, 'test10.txt'))
# f_d = f.decode('utf-8')
i = 1
for line in f:
    # values_d = line.decode('utf-8')
    # values = line.split()
    # line_d = codecs.decode(line, 'utf-8')
    values = line.split()
    word = values[0]
    coefs = np.asarray(values[1:], dtype='float32')
    embeddings_index[word] = coefs
    print(word, coefs)
    i += 1
f.close()

print('\n', embeddings_index['set'])

# print('Found %s word vectors.' % len(embeddings_index))
```
